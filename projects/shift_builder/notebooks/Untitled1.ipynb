```python
import jpholiday
# 日曜日から始まる28日間の日付を生成
start_date = date 
D = pd.date_range(start=start_date, periods=28)

# 祝日判定のマスク（True=祝日, False=祝日以外） --- jpholiday は要インストール
is_holiday_mask = D.map(jpholiday.is_holiday)

# 曜日判定用に、Pandas の weekday 属性を利用
weekday_series = D.weekday  # Returns an ndarray of 0=Mon ... 6=Sun

# ① 祝日だけを抽出（DatetimeIndex のまま）
D_holiday = D[is_holiday_mask]

# ② 平日だけを抽出
#    - まず「祝日でない」こと
#    - かつ weekday が 0〜4 の範囲
mask_weekday = (~is_holiday_mask) & (weekday_series < 5)
D_weekday = D[mask_weekday]

# ③ 土曜日だけを抽出
mask_saturday = (~is_holiday_mask) & (weekday_series == 5)
D_saturday = D[mask_saturday]

# ④ 日曜日だけを抽出
mask_sunday = (~is_holiday_mask) & (weekday_series == 6)
D_sunday = D[mask_sunday]

# 結果を確認
print("元のDatetimeIndex D:")
print(D, "\n")

print("平日の DatetimeIndex:")
print(D_weekday, "\n")

print("土曜の DatetimeIndex:")
print(D_saturday, "\n")

print("日曜の DatetimeIndex:")
print(D_sunday, "\n")

print("祝日の DatetimeIndex:")
print(D_holiday, "\n")
```

```python
# ———— 事前に定義しておくコードリスト ————
pre_list_rest    = R        # 休暇として扱うコード
pre_list_work    = K        # 勤務として扱うコード
pre_list_special = ['廃休', 'マル超']

# ———— 社員名リスト ————
sht   = wb.sheets['Sheet1']
names = sht.range('B23').expand('down').value  # B23, B24, … に社員名

# ———— 辞書の初期化 ————
pre_dict_rest    = {}  # {(社員名, Timestamp): 休暇コード}
pre_dict_work    = {}  # {(社員名, Timestamp, 勤務コード): 勤務コード}
pre_dict_special = {}  # {(社員名, Timestamp): 特別業務コード}
pre_dict_time    = {}  # {(社員名, Timestamp): 時間指定などの値}

# ———— ループしてセルをチェック ————
for i, name in enumerate(names):
    if not name:
        continue
    row = 23 + i
    for j, date in enumerate(D):
        col = 3 + j
        val = sht.range((row, col)).value
        # 空セル or nan はスキップ
        if val is None or (isinstance(val, float) and pd.isna(val)) or val == '':
            continue

        # --- ここで必ず pd.Timestamp に変換 ---
        # 文字列、datetime.date, datetime.datetime, pd.Timestamp の
        # いずれも pd.Timestamp に統一
        date_key = pd.to_datetime(date).normalize()

        # 休暇コードなら pre_dict_rest
        if val in pre_list_rest:
            pre_dict_rest[(name, date_key)] = val

        # 勤務コードなら pre_dict_work
        elif val in pre_list_work:
            pre_dict_work[(name, date_key)] = val

        # “廃休”または“マル超”なら pre_dict_special に登録
        elif val in pre_list_special:
            pre_dict_special[(name, date_key)] = val

        else:
            # それ以外は「時間指定」用として pre_dict_time に登録
            pre_dict_time[(name, date_key)] = val

# ———— 確認出力（デバッグ） ————
import pprint
print("■ 事前指定休暇")
pprint.pprint(pre_dict_rest)
print("■ 事前指定勤務")
pprint.pprint(pre_dict_work)
print("■ 事前指定特別業務")
pprint.pprint(pre_dict_special)
print("■ 事前指定時間")
pprint.pprint(pre_dict_time)
```

```python
# 1) Python の weekday() から曜日ラベルへのマッピング
weekday_map = {0: '月', 1: '火', 2: '水', 3: '木', 4: '金', 5: '土', 6: '日'}

# 2) 各日付 d∈D を「月～日 or 祝」に変換
date_label = {}
for d in D:
    if jpholiday.is_holiday(d):      # 祝日なら '祝'
        date_label[d] = '祝'
    else:
        date_label[d] = weekday_map[d.weekday()]
#各社員が出勤できる曜日を1、出勤しない(ソルバーが選ばない曜日)を0に設定
# 3) avail[(i,d)] = 1 or 0 の辞書を作成　
#事前指定で出勤しない日に指定が入っていたら1にする
avail = {
    (i, d): 1 if (i, d) in pre_dict_work else C[i][ date_label[d] ]
    for i in C
    for d in D
}
```

```python
req = {}  # {(pd.Timestamp, 業務種類): 1}

for d in D:
    # d は pd.Timestamp
    # 祝日判定には d.date()（datetime.date）を渡す
    if jpholiday.is_holiday(d.date()):
        label = '祝'
    else:
        wd = d.weekday()  # 0=月…6=日
        if wd < 5:
            label = ['月','火','水','木','金'][wd]
        else:
            label = '土' if wd == 5 else '日'

    # mapping[label] の各業務種類をすべて需要１として登録
    for k in mapping.get(label, []):
        req[(d, k)] = 1
```

```python
#数理モデルのインスタンス作成
model = pulp.LpProblem("Shift_Scheduling", pulp.LpMinimize)
```

```python
# バイナリ変数の定義
x = pulp.LpVariable.dicts(
    "Shift",
    [
        (i, d, k)
        for i in I
        for d in D
        for k in K
    ],
    cat='Binary'
)
```

```python
# 補助変数 y[i,d] = 1 なら社員iが日にちdに勤務
y = pulp.LpVariable.dicts(
    "WorkDay",
    [(i, d) for i in I for d in D],
    cat='Binary'
)
```

```python
# 休みタイプ変数 rest[i,d,r]　必要
rest = pulp.LpVariable.dicts(
    "Rest",
    [(i, d, r) for i in I for d in D for r in R],
    cat='Binary'
)
```

```python
specialWork = pulp.LpVariable.dicts(
    "X_Special",
    [
        (i,d,s)
        for i in I
        for d in D
        for s in special_attendance  # (廃休 or マル超)
    ],
    cat=pulp.LpBinary
)
```

```python
#欠区の定義
# missing=1 なら (d,t,k) は欠区（割り当て0人）
# missing=0 なら 需要をフルに満たす
missing = pulp.LpVariable.dicts(
    "Missing",
    [(d,k) for (d,k) in req.keys()],
    cat='Binary'
)
```

```python
#連勤を判定するための補助変数
normal_work = pulp.LpVariable.dicts(
    "NormalWork",
    [(i,d) for i in I for d in D],
    cat=pulp.LpBinary
)
```

```python
#曜日による出勤の指定
#事前指定より先に実行して、事前指定があれば上書きされるようにする
for i in I:
    for d in D:
        for k in K:
            # avail[(i,d)] が 0(出勤しない曜日) のとき x[(i,d,k)] == 0
            model += x[(i, d, k)] <= avail[(i, d)], f"Avail_{i}_{d}_{k}"
```

```python
# ---------- 休暇の事前指定 ----------
for i in I:
    for d in D:
        # ----- 休暇指定がある場合 (事前指定で休暇にする日) -----
        if (i, d) in pre_dict_rest:
            rest_type = pre_dict_rest[(i, d)]
            model += rest[(i, d, rest_type)] == 1, f"Rest_{i}_{d}_{rest_type}"
           # 他の通常休暇は0
            for r_ in R:
                if r_ != rest_type:
                    model += rest[(i, d, r_)] == 0, f"RestZero_{i}_{d}_{r_}" 

            # 休暇指定の日は勤務しない → x=0, specialWork=0
            for k in K:
                model += x[(i, d, k)] == 0, f"ZeroX_{i}_{d}_{k}"
                
            # 特別状態(廃休,マル超) も0
            for s in special_attendance:
                model += specialWork[(i, d, s)] == 0, f"ZeroSpWork_{i}_{d}_{s}"

        else:
            # ----- 休暇指定がない → 勤務の事前指定をチェック ----
            # 休暇指定がない
            # ソルバーに選ばせたくない休暇 (例: "計年","夏休"...) は 0固定
            # 逆に "週休","非番","祝休"などはソルバーが使うかもしれない(要件次第)
            for r in R:
                if r not in ["週休", "非番", "祝休"]:
                    model += rest[(i, d, r)] == 0, f"ZeroRest_{i}_{d}_{r}"
```

```python
#"""廃休、マル超の事前指定"""
for i in I:
    for d in D:
        if (i, d) in pre_dict_special:
            s_type = pre_dict_special[(i, d)]
            for s in special_attendance:
                if s == s_type:
                    model += specialWork[(i, d, s)] == 1, f"FixSpec_{i}_{d}_{s}"
                else:
                    model += specialWork[(i, d, s)] == 0, f"ZeroSpec_{i}_{d}_{s}"

        else:
            # 特別状態の指定なし -> 全0
            for s in special_attendance:
                model += specialWork[(i, d, s)] == 0, f"NoSpec_{i}_{d}_{s}"
```

```python
#"""業務の事前指定"""
for i in I:
    for d in D:
        for k in K:
            # もし (i, d) が pre_dict_work にあって、
            # その値がこの k なら x[(i,d,k)] == 1
            if pre_dict_work.get((i, d)) == k:
                model += x[(i, d, k)] == 1, f"ForceWork_{i}_{d}_{k}"
            else:
                # req に無いなら 0 固定
                if (d, k) not in req:
                    model += x[(i, d, k)] == 0, f"NoReq_{i}_{d}_{k}"
                else:
                    # それ以外はソルバー自由
                    pass
```

```python
# 各社員が各日にちに勤務(y[i,d])するか、休暇(rest[i,d,r])を取るか、
for i in I:
    for d in D:
        model += (
            y[i, d]
            + pulp.lpSum(rest[i, d, r] for r in R)
        ) == 1, f"WorkOrRest_{i}_{d}"
```

```python
#"""１日に入れるのは１シフトのみ"""
for i in I:
    for d in D:
        # シフト×業務 の合計
        sum_x_id = pulp.lpSum(x[(i, d, k)] for k in K)

        # 1) 勤務するなら sum_x_id = 1
        # 2) 勤務しない(=0)なら sum_x_id = 0
        #  → sum_x_id == y[i,d]　1日出勤フラグの紐付け(重要)
        model += sum_x_id == y[(i,d)], f"OneShiftOrNone_{i}_{d}"
```

```python
for i in I:
    for d in D:
        for k in K:
            # 1) A に含まれる業務なら、ソルバー自由
            #    ただし (i,d) に対する事前指定がこの k なら強制 x=1
            if k in A[i]:
                if pre_dict_work.get((i, d)) == k:
                    model += x[(i, d, k)] == 1, f"MustWork_{i}_{d}_{k}"
                else:
                    # ソルバー自由 => 制約は書かない
                    pass

            # 2) B に含まれる業務なら、事前指定がこの k なら x=1、されていなければ x=0
            elif k in B[i]:
                if pre_dict_work.get((i, d)) == k:
                    model += x[(i, d, k)] == 1, f"MustWork_{i}_{d}_{k}"
                else:
                    model += x[(i, d, k)] == 0, f"NoWork_{i}_{d}_{k}"

            # 3) どちらにもない業務は担当不可
            else:
                model += x[(i, d, k)] == 0, f"InvalidJob_{i}_{d}_{k}"
```

```python
for i in I:
    for d in D:
        # 週休 + 廃休 <= 1
        model += rest[(i,d,'週休')] + specialWork[(i,d,'廃休')] <= 1
        # 非番 + マル超 <= 1
        model += rest[(i,d,'非番')] + specialWork[(i,d,'マル超')] <= 1
```

```python
for i in I:
    # 週休は月に4日
    model += (
        pulp.lpSum(
            rest[(i,d,'週休')] + specialWork[(i,d,'廃休')]
            for d in D
        ) == 4
    ), f"WeeklyOff_{i}"

    # 非番は4日以上
    model += (
        pulp.lpSum(
            rest[(i,d,'非番')] + specialWork[(i,d,'マル超')]
            for d in D
        ) >= 4
    ), f"Hiban_{i}"
```

```python
#週に1回週休をとる(廃休をとってもカウントされる)
for i in I:
    for w, days_in_week in enumerate(W):
        #rest(週休) + x_special(廃休) == 1
        model += (
            pulp.lpSum(rest[(i,d,'週休')] + specialWork[(i,d,'廃休')]
                        for d in days_in_week) == 1
        )
```

```python
#スラック変数
devPos = {}
devNeg = {}
```

```python
#なるべく週に１回非番をとるがとれない場合はスラック変数により別の週にとることも可
for i in I:
    for w, days_in_week in enumerate(W):
        # 週の "(非番 + マル超)" 回数
        sumHibanOrMarucho = pulp.lpSum(
            rest[(i,d,'非番')] + specialWork[(i,d,'マル超')]
            for d in days_in_week
        )
        
        # スラック変数
        devPos[(i,w)] = pulp.LpVariable(f"devPos_{i}_{w}", lowBound=0)
        devNeg[(i,w)] = pulp.LpVariable(f"devNeg_{i}_{w}", lowBound=0)
        
        model += sumHibanOrMarucho - 1 <= devPos[(i,w)]
        model += 1 - sumHibanOrMarucho <= devNeg[(i,w)]
```

```python
#日曜日に休む場合は週休
for i in I:
    for d in D_sunday:
        model += y[i,d] + rest[i,d,'週休'] == 1, f"SundayWorkOrWeeklyOff_{i}_{d}"
```

```python
#土曜日に休む場合は非番
for i in I:
    for d in D_saturday:
        # もし (i,d) に特別休暇が指定されていればスキップ
        if (i, d) in pre_dict_rest:
            continue
        
        # 指定の休みが無い => 土曜は (出勤 or 非番)
        model += (
            y[(i,d)] + rest[(i,d,'非番')]
            == 1
        ), f"SaturdayWorkOrNonban_{i}_{d}"
```

```python
#祝日以外の日は祝休を取れない
for i in I:
    for d in D:
        if d not in D_holiday:
            # 祝日以外の日 => 祝日休 は 0
            model += rest[i,d,'祝休'] == 0, f"NoHolidayRest_{i}_{d}"
```

```python
# 祝日に非番の指定
for d in D_holiday:
    for i in I:
        #非番の指定があれば入れられる。
        if (i,d) in pre_dict_rest and pre_dict_rest[(i,d)] == '非番':
            continue
        #指定がなければ祝日は "出勤 or 祝休" のどちらか
        else:
            model += y[(i,d)] + rest[(i,d,'祝休')] == 1
```

```python
#"""連勤に関する制約"""
#10連勤の禁止　廃休、マル超を取れば９連勤まで可
for i in I:
    # d_idxを 0..(28-10) までループ => 0..18
    for d_idx in range(len(D) - 10 + 1):  # 28-10+1=19
        # d_idx..d_idx+9 (10日分)
        consecutive_days = [D[d_idx + offset] for offset in range(10)]
        
        # y[(i, day)] を連続10日分集計
        model += (
            pulp.lpSum(y[(i, day_)] for day_ in consecutive_days)
            <= 9
        ), f"No10ConsecutiveDays_{i}_{d_idx}"
                
#紐付け制約: normal_work <= y, normal_work <= 1 - x_special
for i in I:
    for d in D:
        model += normal_work[(i,d)] >= y[(i,d)] - specialWork[(i,d,'廃休')] - specialWork[(i,d,'マル超')]
        model += normal_work[(i,d)] <= 1 - specialWork[(i,d,'廃休')]
        model += normal_work[(i,d)] <= 1 - specialWork[(i,d,'マル超')]
        

#廃休、マル超を取らない連勤は５まで。
for i in I:
    for start_idx in range(len(D) - 6 + 1):  # 0..22
        consecutive_dates = [D[start_idx + r] for r in range(6)]
        # ここで normal_work[(i, consecutive_dates[r])] を合計
        model += (
            pulp.lpSum(normal_work[(i, d_)] for d_ in consecutive_dates)
            <= 5
        )
```

```python
#"""社員の出勤可能数と業務数の過不足を計算"""
# 1) 社員ごとの出勤可能日数を計算
# 1) 社員ごとの出勤可能日数を、C の可否から直接カウント
base = 20  # 最大勤務日数の上限

# avail[(i, d)] が 1 の日を数えて possible を計算し、上限をかける
employee_possible_days = {
    i: min(
        sum(avail[(i, d)] for d in D),
        base
    )
    for i in I
}

# 確認出力
for i, cnt in employee_possible_days.items():
    print(f"{i} の出勤可能日数: {cnt} 日")
for i in I:
    extra_rest_count = 0
        
    # (1-a) 特別休暇を差し引く
    for d in D:
        if (i, d) in pre_dict_rest:
            r_type = pre_dict_rest[(i, d)]
            if r_type in ['計年', '年休', '夏期', '冬期', '代休',
                            '承欠', '休職', '産休', '育休', '介護', '病休', 'その他']:
                extra_rest_count += 1
    # 仮での基本出勤可能日数
    employee_possible_days[i] -= extra_rest_count

    # (1-b) 廃休 or マル超 指定があれば +1
    for d in D:
        if (i, d) in pre_dict_special:
            s_type = pre_dict_special[(i,d)]
            if s_type in ['廃休','マル超']:
                employee_possible_days[i] += 1

# (1-c) 出張や研修など (reqにない業務) の事前指定があれば -1
for (i, d), k in pre_dict_work.items():
    if (d, k) not in req:
        # この業務に入ってもreqは満たせないため-1
        employee_possible_days[i] -= 1

# 2) 需要計算
# 2) 需要計算：req による実際の要出勤数を集計
#バランス調整用の補助20を引く
total_demand = sum(req.values()) - 20

print(f"【需要計算】total_demand = {total_demand} シフト枠")

#3) 供給計算
# 例: 全社員の出勤可能日数を合計
total_supply = sum(employee_possible_days.values())

# 4) 過不足を計算
excess_or_deficiency = total_supply - total_demand
print(f"【出勤可能日数計算】")
for i in I:
    print(f"  - {i}: {employee_possible_days[i]}日")

print(f"total_supply={total_supply}")
# あるいはゼロを +0 と表示したくない場合は条件分岐で制御
sign = '+' if excess_or_deficiency > 0 else ''
print(f"過不足 = {sign}{excess_or_deficiency}")
```

```python
def generate_features(df, dates, feature_columns):
    idx    = df.index.union(dates)
    df_ext = df.reindex(idx)

    # 曜日系
    df_ext['dow']     = df_ext.index.dayofweek
    df_ext['dow_sin'] = np.sin(2*np.pi*df_ext['dow']/7)
    df_ext['dow_cos'] = np.cos(2*np.pi*df_ext['dow']/7)

    # 祝日系
    df_ext['is_holiday']             = df_ext.index.to_series().apply(jpholiday.is_holiday).astype(int)
    df_ext['is_after_holiday']       = df_ext['is_holiday'].shift(1).fillna(0).astype(int)
    df_ext['is_after_after_holiday'] = df_ext['is_holiday'].shift(2).fillna(0).astype(int)

    # 月・年末年始・お盆
    df_ext['month']       = df_ext.index.month
    df_ext['season'] = df_ext.index.map(assign_season)
    df_ext['is_new_year'] = ((df_ext.index.month == 1) & (df_ext.index.day <= 3)).astype(int)
    df_ext['is_obon']     = (
        (df_ext.index.month == 8) &
        (df_ext.index.day   >= 13) &
        (df_ext.index.day   <= 16)
    ).astype(int)

    # ラグ・移動平均
    df_ext['lag_1']          = df_ext['通常物数'].shift(1)
    df_ext['lag_7']          = df_ext['通常物数'].shift(7)
    df_ext['rolling_mean_7'] = df_ext['通常物数'].shift(1).rolling(7).mean()

    # 価格増加フラグ
    df_ext['price_increase_flag'] = 0
    df_ext.loc[df_ext.index >= pd.Timestamp('2024-10-01'), 'price_increase_flag'] = 1

    # 最終的に欲しい行・列を抽出
    return df_ext.loc[dates, feature_columns]
```

```python
def assign_season(date):
    if date.month in [3, 4, 5]:
        return 1  # 春
    elif date.month in [6, 7, 8]:
        return 2  # 夏
    elif date.month in [9, 10, 11]:
        return 3  # 秋
    else:
        return 4  # 冬
```

```python
start_date = D[0]  # 予測を始めたい日付
print(start_date)
n_days = 28

# 3) 再帰用の履歴 Series をコピーして用意
#    history は pd.Series で index が日付、値が実測 or 予測の「通常物数」
history = history_df['通常物数'].copy() 
history_rec = history.copy()

feature_columns = [
    'dow_sin','dow_cos','is_holiday','is_after_holiday','is_after_after_holiday',
    'month','season','lag_1','lag_7','rolling_mean_7',
    'is_new_year','is_obon','price_increase_flag'
]

# --- 3) 繰り越し用バッファと予測用Series準備 -----------------------
carry = 0.0
future_dates = [start_date + pd.Timedelta(days=i) for i in range(n_days)]
preds = pd.Series(index=future_dates, dtype=float)

# --- 4) 再帰的予測＋土日祝繰り越し処理 ---------------------------
for dt in future_dates:
    # 特徴量生成
    X_dt = generate_features(
        history.to_frame(name='通常物数'),
        [dt],
        feature_columns
    )
    dmat = xgb.DMatrix(X_dt)

    # 生予測値
    raw = float(bst.predict(dmat)[0])

    # 土日祝か判定
    is_wh = jpholiday.is_holiday(dt) or dt.weekday() >= 5
    if is_wh:
        carry += raw
        pred = 0.0
    else:
        pred = raw + carry
        carry = 0.0

    preds[dt] = pred
    history.loc[dt] = pred  # 次日のラグ用に追加

# --- 5) 結果表示 ----------------------------------------------------------
for dt, v in preds.round(-3).astype(int).items():
    wd = ['月','火','水','木','金','土','日'][dt.weekday()]
    mark = '祝' if jpholiday.is_holiday(dt) else ''
    print(f"{dt:%Y-%m-%d} ({wd}{mark}): {v} 通")
```

```python
# 1) カバレッジ制約：事前指定分を除外して残りを埋める
for (d, k), needed in req.items():
    # その (d,k) に pre_dict_work で強制された社員リスト
    forced_workers = [i for i in I if pre_dict_work.get((i, d)) == k]
    # forced の人数は定数
    forced_count   = len(forced_workers)

    # 強制されていない社員だけを free_workers に
    free_workers = [i for i in I if i not in forced_workers]

    # シンボリックに missing 変数を使う
    #  左辺: free_workers の x の合計 + forced_count
    #  右辺: needed * (1 - missing[(d, k)])
    model += (
        pulp.lpSum(x[(i, d, k)] for i in free_workers) 
        + forced_count
        == needed * (1 - missing[(d, k)])
    ), f"Coverage_{d.date()}_{k}"
```

```python
"""
# 目的関数
    欠区と週1非番ズレを最小化する
"""
# ペナルティ重みづけパラメータ
alpha = 0.1  # alpha を大きくするとペナルティを重視、小さくすると希望度を重視した解になります。

# Combined objective を最小化問題としてセット
model.setObjective(
    # 1) 欠区のペナルティ
    pulp.lpSum(
        priority_map.get((d, k), 0) * missing[(d, k)]
        for (d, k) in req.keys()
    )
    # 2) 週1非番ズレのペナルティ
    + pulp.lpSum(
        devPos[(i, w)] + devNeg[(i, w)]
        for i in I
        for w in range(len(W))
    )

    # 3) 満足度（優先度×割当）をマイナス項で追加
    - alpha * pulp.lpSum(
        A[i][k] * x[(i, d, k)]
        for i in A
        for d in D
        for k in A[i].keys()
    )

)
```
